```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
# df = pd.read_csv('used_car_data.csv')  # Adjust the path as needed

# Inspect the DataFrame
print(df.info())  # Check for non-numeric columns

# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Generate correlation matrix
correlation = numeric_df.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm')

# Title and display
plt.title('Correlation Heatmap of Used Car Features')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Mileage', y='Price')
plt.title('Price vs. Mileage of Used Cars')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
avg_price_by_year = df.groupby('Year')['Price'].mean().reset_index()
sns.lineplot(data=avg_price_by_year, x='Year', y='Price')
plt.title('Average Used Car Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```
